```python
import nltk #Importing and downloading respective nltk and punkt modules
nltk.download('punkt')

def highstr(words):
    count={}
    for i in words:
      if i in count:
        count[i]=count[i]+1
      else:
        count[i]=1
    import operator

#Finding the top 30 words in my respective novel
    top=(dict(sorted(count.items(),key=operator.itemgetter(1), reverse=True)[:30]))
    import pandas as pd

#Creation of Dataframe containing top words
    df = pd.DataFrame.from_dict(top.items())
    df.columns = ['Word', 'Count']
    return (df)

   

#Clean Text function
def cleantext(filename):

#Load Data
    file = open(filename, 'rt',encoding="utf8")
    text = file.read()
    file.close()
#Split text into words
    from nltk.tokenize import word_tokenize
    tokens = word_tokenize(text)
#Converting all letters to lowercase
    tokens = [w.lower() for w in tokens]
#Removal of punctuation 
    import string
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]

#Removal of non-alphabetic characters / numbers
    words = [word for word in stripped if word.isalpha()]

#Flitering of Stop Words
    nltk.download('stopwords')
    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    return words

#Output
def topwords(f):
    return (highstr(cleantext(f)))
```

```python
clean = cleantext ("A_Study_of_Shakespeare.txt") #Creating new variable clean which is clean text version of my novel

count = {}  #saving count of each word
for w in (clean):
    
    if w in count:
        count[w] += 1
    else:
        count[w] = 1
for word, times in count.items():
    print ("The word '%s' was found %d times" % (word, times)) #Output
```
